```python
import random
from sympy import isprime, mod_inverse

def generar_primo_grande(bits):
    # Genera un número primo aleatorio del tamaño especificado en bits
    while True:
        p = random.getrandbits(bits)
        if isprime(p):
            return p

def generar_par_claves(bits):
    p = generar_primo_grande(bits)
    q = generar_primo_grande(bits)
    n = p * q
    phi = (p - 1) * (q - 1)

    # Elige un entero e tal que 1 < e < phi(n) y GCD(e, phi(n)) = 1
    e = random.randrange(1, phi)
    while gcd(e, phi) != 1:
        e = random.randrange(1, phi)

    # Calcula d para satisfacer la relación de congruencia d*e ≡ 1 (mod phi)
    d = mod_inverse(e, phi)

    # Clave pública (e, n) y clave privada (d, n)
    return ((e, n), (d, n))

def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def encriptar(clave_publica, mensaje_claro):
    clave, n = clave_publica
    cifrado = [pow(ord(char), clave, n) for char in mensaje_claro]
    return cifrado

def desencriptar(clave_privada, mensaje_cifrado):
    clave, n = clave_privada
    claro = [chr(pow(char, clave, n)) for char in mensaje_cifrado]
    return ''.join(claro)

# Prueba la implementación de RSA
bits = 4000
clave_publica, clave_privada = generar_par_claves(bits)
print("Clave Pública: ", clave_publica)
print("Clave Privada: ", clave_privada)

mensaje = "hola mundo"
mensaje_cifrado = encriptar(clave_publica, mensaje)
print("Mensaje Encriptado: ", mensaje_cifrado)

mensaje_desencriptado = desencriptar(clave_privada, mensaje_cifrado)
print("Mensaje Desencriptado: ", mensaje_desencriptado)
```
