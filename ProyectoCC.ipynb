```python
import random
from sympy import isprime, mod_inverse

def generate_large_prime(bits):
    # Genera un número primo aleatorio del tamaño especificado en bits
    while True:
        p = random.getrandbits(bits)
        if isprime(p):
            return p

def generate_keypair(bits):
    p = generate_large_prime(bits)
    q = generate_large_prime(bits)
    n = p * q
    phi = (p - 1) * (q - 1)

    # Elige un entero e tal que 1 < e < phi(n) y GCD(e, phi(n)) = 1
    e = random.randrange(1, phi)
    while gcd(e, phi) != 1:
        e = random.randrange(1, phi)

    # Calcula d para satisfacer la relación de congruencia d*e ≡ 1 (mod phi)
    d = mod_inverse(e, phi)

    # Clave pública (e, n) y clave privada (d, n)
    return ((e, n), (d, n))

def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def encrypt(public_key, plaintext):
    key, n = public_key
    cipher = [pow(ord(char), key, n) for char in plaintext]
    return cipher

def decrypt(private_key, ciphertext):
    key, n = private_key
    plain = [chr(pow(char, key, n)) for char in ciphertext]
    return ''.join(plain)

# Prueba la implementación de RSA
bits = 3000
public_key, private_key = generate_keypair(bits)
print("Clave Pública: ", public_key)
print("Clave Privada: ", private_key)

message = "hello world"
encrypted_msg = encrypt(public_key, message)
print("Mensaje Encriptado: ", encrypted_msg)

decrypted_msg = decrypt(private_key, encrypted_msg)
print("Mensaje Desencriptado: ", decrypted_msg)
```
